```python
advanced_features = [

'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode','condition', 'grade', 'waterfront', 'view', 'sqft_above', # square feet above ground 

'sqft_basement', 'yr_built', ]
```

```python
train_data,test_data = home_data.random_split(.8,seed=0)
```

```python
advanced_features_model = graphlab.linear_regression.create(train_data,target='price',features=advanced_features,
                                                            validation_set=None)
```
